First, define the hypothesis $h$

```python
def h(theta_0, theta_1, x):
    return theta_0 + (theta_1 * x)
```

Then, define a list of input values for the hypothesis

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
X = np.arange(-10, 10)
```

Define values for $\Theta_0$ and $\Theta_1$ parameters:

```python
theta_0 = 0
theta_1 = 1
```

Finally, evaluate the function using all of the sample values and defined parameters:

```python
vh = np.vectorize(h)
Y = vh(theta_0, theta_1, X)
```

Plotting the values, it shows the following:

```python
#% matplotlib notebook
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

def plot_data(X, Y, theta_0, theta_1):
    plt.figure()
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(X,Y, label=r'$\Theta_0$ = {}, $\Theta_1$ = {}'.format(theta_0, theta_1))
    plt.title(r'$H(\Theta_0,\Theta_1,\mathbf{X})$')
    plt.legend()
    plt.show()

plot_data(X, Y, theta_0, theta_1)
```

Doing with other values:

```python
# Theta_0 = np.arange(-6,6)
# Theta_1 = np.arange(-6,6)

Theta_0 = np.array([0])
Theta_1 = np.array([-1,-0.5,0,0.5,1])

plt.figure()
plt.xlim(-6, 6)
plt.ylim(-6, 6)
for i in range(0,len(Theta_0)):
    for j in range(0, len(Theta_1)):
        Y = vh(Theta_0[i], Theta_1[j], X)
        # plt.gca().set_aspect('equal', adjustable='box')
        plt.plot(X,Y, label=r'$\Theta_0$ = {}, $\Theta_1$ = {}'.format(Theta_0[i], Theta_1[j]))

plt.legend()
plt.show()
```

So, as we can see, positive values for $\Theta_1$ specify a positive slop $m$. In trigonometry, a a positive slope means an angle that opens clock-counterwise, while a negative slop means an angle that opens clock-wise.
